```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('merged_data.csv')
df = df.drop(columns=['Unnamed: 0'],axis=1)
df
```

```python
# Statistical analysis
statistics = df.describe()
statistics
```

```python
countries = df.Country.to_list()
coffee_consumption_rank = df['Rank By Coffee Consumption'].to_list()
death_rate_rank = df['Rank By liver Death Rate'].to_list()

plt.figure(figsize=(12, 6))

# Coffee consumption rankings
plt.subplot(1, 2, 1)
plt.barh(countries, coffee_consumption_rank, color='skyblue')
plt.xlabel('Ranking')
plt.title('Coffee Consumption Rankings')
plt.gca().invert_yaxis()  # Invert y-axis to have highest rank on top

# Liver death rate rankings
plt.subplot(1, 2, 2)
plt.barh(countries, death_rate_rank, color='salmon')
plt.xlabel('Ranking')
plt.title('Liver Death Rate Rankings')
plt.gca().invert_yaxis()  # Invert y-axis to have highest rank on top

plt.tight_layout()
plt.show()
```

```python

# Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Coffee Consumption (Lbs per Person Per Year)", y="Death Rate", data=df)
plt.title("Scatter Plot of Coffee Consumption vs. Liver Death Rate")
plt.xlabel("Coffee Consumption (Lbs per Person Per Year)")
plt.ylabel("Liver Death Rate")
plt.grid(True)
plt.show()
```

```python

# Box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x="Coffee Consumption (Lbs per Person Per Year)", y="Country", data=df)
plt.title("Box Plot of Coffee Consumption Across Countries")
plt.xlabel("Coffee Consumption (Lbs per Person Per Year)")
plt.ylabel("Country")
plt.grid(True)
plt.show()
```

```python

# Bar plot
plt.figure(figsize=(8,5))
sns.barplot(x="Coffee Consumption (Lbs per Person Per Year)", y="Country", data=df, ci=None)
plt.title("Bar Plot of Coffee Consumption Across Countries")
plt.xlabel("Coffee Consumption (Lbs per Person Per Year)")
plt.ylabel("Country")
plt.grid(True)
plt.show()
```

```python
df_for_corr = df.drop(columns=['Country'],axis=1)
```

```python
df_for_corr.describe()
```

```python
correlation = df_for_corr.corr()

# Heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Coffee Consumption:
- The mean coffee consumption per person per year among the listed countries is approximately 14.34 lbs, with a standard deviation of 7.51 lbs. This indicates a considerable variability in coffee consumption levels across the countries.
- The minimum coffee consumption recorded is 9.26 lbs per person per year, while the maximum is 26.45 lbs. This range suggests diverse coffee consumption habits among the listed countries.
- The interquartile range (IQR), which represents the middle 50% of the data, ranges from approximately 11.13 to 17.57 lbs. This suggests that most countries fall within this range of coffee consumption.

## Liver Death Rate:
- The mean liver death rate among the listed countries is approximately 4.44 per 100,000 population, with a standard deviation of 1.03. This indicates some variability in liver death rates across the countries.
- The minimum liver death rate recorded is 2.59 per 100,000 population, while the maximum is 6.66. This range suggests differences in liver health outcomes among the countries.
- The interquartile range (IQR) for liver death rates ranges from approximately 3.79 to 5.01 per 100,000 population. This indicates that most countries fall within this range of liver death rates.

## Comparison:
- There is no apparent direct relationship between coffee consumption and liver death rates based solely on these summary statistics. However, further analysis, such as correlation analysis and visualizations, may reveal potential associations between these variables.
